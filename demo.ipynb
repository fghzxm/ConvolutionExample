```julia
using Convolution, Plots
```

```julia
# Discrete convolution

discrete_convolved = discrete_convolve(ones(11), ones(7))
```

```julia
# Continuous convolution with different sampling frequencies

f = restrict(x -> 1.0, (-5.5, 5.5))
h = restrict(x -> 1.0, (-3.5, 3.5))

# Convolve with sampling frequency = 10 Hz
convolved = convolve(.1, f, h)
# Convolve with sampling frequency = 1 Hz
coarse_convolved = convolve(1.0, f, h)

nothing
```

```julia
# Plot

plot(-11 : 11, [0, 0, 0, discrete_convolved..., 0, 0, 0], seriestype = :scatter, label = "Discrete")
plot!(-12 : .01 : 12, x -> convolved(x), label = "Fine continuous")
plot!(-12 : .01 : 12, x -> coarse_convolved(x), label = "Coarse continuous")
```
